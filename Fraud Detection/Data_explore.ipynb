# Visualisation data

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
%matplotlib inline
```

```python
data =  pd.read_csv('./Data/creditcard.csv')
```

### How many are frauds and how many are not fraud

```python
class_name = {0: 'Not Fraud', 1: 'Fraud'}
print(data.Class.value_counts().rename(index=class_name))
```

### Plotting variables using a subplot

```python
start_plot = data.columns.get_loc('V1')
stop_plot  = data.columns.get_loc('Class')
```

```python
fig = plt.figure(figsize=(15,12))
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

for i in range(start_plot,stop_plot):
    plt.subplot(5,6,i);
    plt.plot(data.iloc[:,i]);
    plt.subplot(5,6,i).title.set_text(data.columns[i])

plt.show()
```

### Ploting correlation matrix

```python
corr = data.corr()
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
```

```python
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
